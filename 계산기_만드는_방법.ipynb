# 계산하는 기본 지식
1. 변수 사용
2. 리스트 사용
3. 반복문 사용 (for문)

```python
# 변수: 값을 저장하고 바꾸기

value = 0                     # 계산기 초기값
value = value + 10
print(value)

value = value -3
print(value)
```

```python
# 리스트: 여러 값을 한 번에 저장

numbers = [1,2,3]
print(numbers[0])              # 인덱스 값으로 출력
print(numbers[1])
print(numbers[2])
```

```python
# 반복문: 차례대로 처리하기
numbers = [1,2,3]
total = 0

for n in numbers:               # number에 있는 값을 하나씩 꺼내서 n에 넣어주기
    total = total + n
    print(total)                # 0 + 1 =1 / 1 + 2 = 3 / 3 + 3 = 6
```

```python
# 함수: 기능을 묶기

value = 0                       # 전역 변수 만들기

def add(num):                   # add 함수 정의
    global value
    value += num

def sub(num):                   # sub 함수 정의
    global value
    value -= num

add(10)                          # add 함수 정의 후 add 호출하기
sub(3)
print(value)
```

```python
# 클래스: 값  + 기능을 한번에 묶기
# 관습적인 함수 add / sub / mul / div 사용

class Calculator:                      # 계산기 설계도
    def __init__(self):                # 계산기 초기값 설정
        self.value =0

    def add(self, num):                # 더하기 함수 정의
        self.value += num

    def sub(self, num):                # 빼기 함수 정의
        self.value -= num

    def mul(self, num):                # 곱하기 함수 정의
        self.value *= num

    def div(self, num):                 # 나누기 함수 정의
        self.value /= num               # self.value ( 계산기 안에 들어있는 숫자 화면을 의미함 )

c = Calculator()                        # 설계도로 만든 실제 계산기

c.add(10)                               # 계산기의 버튼과 내부 연산 가능
c.sub(4)
c.mul(2)
c.div(7)

print(c.value)
```

```python
# 예제) 2 - 10 + 7 계산하기

class Calculator:
    def __init__(self):
        self.value = 0

    def firstnum(self,num):
        self.value = num

    def sub(self,num):
        self.value -= num

    def add(self,num):
        self.value += num

c = Calculator()

c.firstnum(2)
c.sub(10)
c.add(7)
print(c.value)
```

```python
#예제) 체중 증가

class WeightTracker:
    def __init__(self):
        self.value = 60
    def add1(self,num):
        self.value += num
    def sub(self,num):
        self.value -= num
    def add2(self,num):
        self.value += num

tracker = WeightTracker()

tracker.add1(5)
tracker.sub(3)
tracker.add2(2)

print(tracker.value)

```
