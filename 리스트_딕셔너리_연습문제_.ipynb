```python
학생 = ['영희','철수']
반 = ['1반','2반']
번호 = [1,2,3]
리스트 = []

for i in range(len(학생)):
    for j in 반:
        for k in 번호:
            리스트.append(학생[i] + "-" + j + "-" + str(k) )
print(리스트)
```
